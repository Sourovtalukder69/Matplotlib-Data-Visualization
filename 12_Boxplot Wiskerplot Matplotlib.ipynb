# Boxplot and Wiskerplot

```python
# Box plot: Box plot is a plotting system use to findout outliers in a data. 
#                            BOX AND WHISKER
#               --------[          |         ]---------- o 
#               L       Q1        Mean       Q3        H 
# Q1 = 1st Quartile(25% pecentile)
# Q3 = 3nd Quartile(75% percentile)
# (----) wisker plot
# L = lowest value | H = highest value
# o = outlier
  

# creating a simple box plot of some data
# `width` parameter use for increase or decrease box size
import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50, 60, 70]
plt.boxplot(x)
plt.show()
```

```python
# `notch` parameter use for v shape plot
import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50, 60, 70]
plt.boxplot(x, notch=True)
plt.show()
```

```python
# `vert=False` use for creating horizontal box plot
# `widths` paramter use for change box size 
# `label` parameter use for labelling
#` patch_artist` paramter use for box background change

import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50, 60, 70]
plt.boxplot(x,vert=False, widths=0.2, labels=["python"], patch_artist=True)
plt.show()

```

```python

# `showmeans=True` parameter use for showing mean exact location
# for see outlier inset a outlier value
# `whis` parameter use to join the line with outlier

import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50, 60, 70, 120]
plt.boxplot(x,vert=False, widths=0.2, labels=["python"], showmeans=True, whis= 4)
plt.show()

```

```python

# `sym=" "` parameter use for color the outlier so that visible clearly
# `boxprops=dict(color="")` parameter use for change box properies
# `capprop=dict(color="blue")` use for max min porperty
# `whiskerprops=dict(color="k")`
# `filerprops=dict(color="y")`
import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50, 60, 70, 120]
plt.boxplot(x,vert=False, labels=["python"], showmeans=True,sym="g+", boxprops=dict(color="r"), capprops=dict(color="blue"), whiskerprops=dict(color="k"))
plt.show()

```

```python

# To plot multiple boxplot
import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50, 60, 70]
y = [50, 55, 60, 65, 70, 75, 80]
z = [x, y]
plt.boxplot(z, labels=["plot1", "plot2"])
plt.show()
```
