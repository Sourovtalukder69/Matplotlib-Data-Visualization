```python
iris = sns.load_dataset('iris')
iris
```

```python
iris['species'].replace({'setosa':0, 'virginica':1, 'versicolor':2}, inplace=True)
iris.sample(5)
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', data=iris, hue='species', palette='cool')
```

```python
plt.scatter(iris['sepal_length'],iris['petal_length'],c=iris['species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
plt.figure(figsize=(5, 3))
plt.scatter(iris['sepal_length'], iris['petal_length'], c=iris['species'], cmap='viridis',)
plt.colorbar()
plt.show()
```
